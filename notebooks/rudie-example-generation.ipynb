```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.linalg import block_diag
import scipy.linalg as linalg

from filterpy.kalman import KalmanFilter

np.set_printoptions(precision=17)

state_buffer = np.array(
    [0.040157366134307, -0.004584667101862575, -0.13210335109501573, -0.06167960574363984, -0.09218946743133555, 0.04102534430132926, 0.055414343550211326, 0.023937347653503412, 0.004426223901518564, -0.012629592056939521]
)
state_buffer_ndx = 0

measurement_noise_buffer = np.array(
    [0.040157366134307, -0.004584667101862575, -0.13210335109501573, -0.06167960574363984, -0.09218946743133555, 0.04102534430132926, 0.055414343550211326, 0.023937347653503412, 0.004426223901518564, -0.012629592056939521]
)
measurement_noise_buffer_ndx = 0

process_noise_buffer = np.array(
    [0.0012698874181772126, -0.0001449799035552893, -0.004177474760011482, -0.0019504803933110553, -0.002915286933609328, 0.0012973352978480962, 0.00175235540661729, 0.0007569653972835781, 0.00013996948962675476, -0.00039938276818699854]
)
process_noise_buffer_ndx = 0

kf = KalmanFilter(dim_x=2, dim_z=1)

state = np.array([[state_buffer[state_buffer_ndx], state_buffer[state_buffer_ndx+1]]]).T
state_buffer_ndx += 2

kf.F = np.array([[1., 1.],
                 [0., 1.]])
kf.Q = block_diag(1e-5, 1e-5)

kf.H = np.array([[1., 0]])
kf.R = block_diag(1e-1)

kf.x = np.array([[-0.13210335109501573, -0.06167960574363984]]).T
kf.P = block_diag(1., 1.)

for cycle in range(1, 4):
    kf.predict()
    
    print("kalman spin {}".format(cycle))
    
    print("kf.x")
    print("{}".format(kf.x))
    print("kf.P")
    print("{}".format(kf.P))
    
    print()
    
    measurement = np.dot(kf.H, state) + measurement_noise_buffer[measurement_noise_buffer_ndx]
    measurement_noise_buffer_ndx += 1
    
    kf.update(measurement)
    
    process_noise = np.array([[process_noise_buffer[process_noise_buffer_ndx], process_noise_buffer[process_noise_buffer_ndx+1]]]).T
    process_noise_buffer_ndx += 2
    state = np.dot(kf.F, state) + process_noise
    
    
    print("kf.y: {}".format(kf.y))
    print("kf.S: {}".format(kf.S))
    print("kf.K: {}".format(kf.K))
    print("kf.x: {}".format(kf.x))
    print("kf.P: {}".format(kf.P))
    
    print()
```
